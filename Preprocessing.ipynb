### Traintement compact de train et test 

```python
def preproc1(train,test):
    print('Processing Train data')
    ## Hydrologie
    varh  = train[['Horizontal_Distance_To_Hydrology','Vertical_Distance_To_Hydrology']]
    leh = LabelEncoder()
    h = [np.array2string(i) for i in  varh.values]
    X = le.fit_transform(h)
    Y = [int(np.linalg.norm(i,2)) for i in varh.values]
    train.insert(4,'Hydrology',Y)
    
    ## Shade
    varsh  = train[train.columns[8:11]]
    les = LabelEncoder()
    h = [np.array2string(i) for i in  varsh.values]
    X = les.fit_transform(h)
    train.insert(8,'Shade',X)

    
    ## Wilderness
    varw  = train[train.columns[13:17]]
    lew = LabelEncoder()
    h = [np.array2string(i) for i in  varw.values]
    X = lew.fit_transform(h)
    train.insert(13,'Wilderness',X)
    
    ## Soil
    varsoil = train[train.columns[18:58]]
    lesoil = LabelEncoder()
    h = [np.array2string(i) for i in  varsoil.values]
    X = lesoil.fit_transform(h)
    train.insert(18,'Soil',X)
    
    print('Processing Test data')
    ## On garde les memes encodeurs pour test et train
    ## Hydrologie
    varh  = test[['Horizontal_Distance_To_Hydrology','Vertical_Distance_To_Hydrology']]
    h = [np.array2string(i) for i in  varh.values]
    X = le.fit_transform(h)
    Y = [int(np.linalg.norm(i,2)) for i in varh.values]
    test.insert(4,'Hydrology',Y)
    
    
    ## Shade
    varsh  = test[test.columns[8:11]]
    h = [np.array2string(i) for i in  varsh.values]
    X = les.fit_transform(h)
    test.insert(8,'Shade',X)

    
    ## Wilderness
    varw  = test[test.columns[13:17]]
    h = [np.array2string(i) for i in  varw.values]
    X = lew.fit_transform(h)
    test.insert(13,'Wilderness',X)
    
    ## Soil
    varsoil = test[test.columns[18:58]]
    h = [np.array2string(i) for i in  varsoil.values]
    X = lesoil.fit_transform(h)
    test.insert(18,'Soil',X)
```

```python
classifier = SVC(C=10, # penalty parameter
                 kernel='rbf', # kernel type, rbf working fine here
                 degree=3, # default value
                 gamma=1, # kernel coefficient
                 coef0=1, # change to 1 from default value of 0.0
                 shrinking=True, # using shrinking heuristics
                 tol=0.001, # stopping criterion tolerance 
                 probability=False, # no need to enable probability estimates
                 cache_size=200, # 200 MB cache size
                 class_weight=None, # all classes are treated equally 
                 verbose=False, # print the logs 
                 max_iter=-1, # no limit, let it run
                 decision_function_shape=None, # will use one vs rest explicitly 
                 random_state=None)
model = OneVsRestClassifier(classifier, n_jobs=-1)
```

## Crossvalidation

```python
classifier = SVC(C=38, # penalty parameter
                 kernel='rbf', # kernel type, rbf working fine here
                 degree=3, # default value
                 gamma=1, # kernel coefficient
                 coef0=1, # change to 1 from default value of 0.0
                 shrinking=True, # using shrinking heuristics
                 tol=0.001, # stopping criterion tolerance 
                 probability=False, # no need to enable probability estimates
                 cache_size=200, # 200 MB cache size
                 class_weight=None, # all classes are treated equally 
                 verbose=False, # print the logs 
                 max_iter=-1, # no limit, let it run
                 decision_function_shape=None, # will use one vs rest explicitly 
                 random_state=None)
model = OneVsRestClassifier(classifier, n_jobs=1)
```

```python
parameters = {"estimator__gamma":(uniform(.0000005,.5))}
rdgrid_search = RandomizedSearchCV(model, parameters,n_iter=3, n_jobs=-1)
```

```python
%%time
rdgrid_search.fit(X,y)
```

```python
pd.DataFrame(rdgrid_search.cv_results_)[['mean_test_score', 'std_test_score', 'params']]
```

```python
g = np.power([.1],[9,8,7])
parameters = {"estimator__gamma":uniform(1e-9,9.99999e-9)}
grid_search = RandomizedSearchCV(model, parameters,n_jobs = -1,n_iter = 20)
```

```python
%%time
grid_search.fit(X, y)
```

```python
pd.set_option('display.max_colwidth', -1)
pd.DataFrame(grid_search.cv_results_)[['mean_test_score', 'std_test_score', 'params']]
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
g=[i['estimator__gamma'] for i in grid_search.cv_results_['params']]
ax.semilogx(g,pd.DataFrame(grid_search.cv_results_)[['mean_test_score']],'.')
plt.show()
```
