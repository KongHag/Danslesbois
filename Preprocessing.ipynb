### Traintement compact de train et test 

```python
def preproc1(train,test):
    print('Processing Train data')
    ## Hydrologie
    varh  = train[['Horizontal_Distance_To_Hydrology','Vertical_Distance_To_Hydrology']]
    leh = LabelEncoder()
    h = [np.array2string(i) for i in  varh.values]
    X = le.fit_transform(h)
    train.insert(4,'Hydrology',X)
    
    ## Shade
    varsh  = train[train.columns[8:11]]
    les = LabelEncoder()
    h = [np.array2string(i) for i in  varsh.values]
    X = les.fit_transform(h)
    train.insert(8,'Shade',X)

    
    ## Wilderness
    varw  = train[train.columns[13:17]]
    lew = LabelEncoder()
    h = [np.array2string(i) for i in  varw.values]
    X = lew.fit_transform(h)
    train.insert(13,'Wilderness',X)
    
    ## Soil
    varsoil = train[train.columns[18:58]]
    lesoil = LabelEncoder()
    h = [np.array2string(i) for i in  varsoil.values]
    X = lesoil.fit_transform(h)
    train.insert(18,'Soil',X)
    
    print('Processing Test data')
    ## On garde les memes encodeurs pour test et train
    ## Hydrologie
    varh  = test[['Horizontal_Distance_To_Hydrology','Vertical_Distance_To_Hydrology']]
    h = [np.array2string(i) for i in  varh.values]
    X = le.fit_transform(h)
    test.insert(4,'Hydrology',X)
    
    ## Shade
    varsh  = test[test.columns[8:11]]
    h = [np.array2string(i) for i in  varsh.values]
    X = les.fit_transform(h)
    test.insert(8,'Shade',X)

    
    ## Wilderness
    varw  = test[test.columns[13:17]]
    h = [np.array2string(i) for i in  varw.values]
    X = lew.fit_transform(h)
    test.insert(13,'Wilderness',X)
    
    ## Soil
    varsoil = test[test.columns[18:58]]
    h = [np.array2string(i) for i in  varsoil.values]
    X = lesoil.fit_transform(h)
    test.insert(18,'Soil',X)
```

```python

train = pd.read_csv(r'C:\Users\Tim\Desktop\tensorflow\Dans les bois\Danslesbois\train.csv',sep=',')
test = pd.read_csv(r'C:\Users\Tim\Desktop\tensorflow\Dans les bois\Danslesbois\test.csv',sep=',')
```

```python
%%time
preproc1(train,test )
```
